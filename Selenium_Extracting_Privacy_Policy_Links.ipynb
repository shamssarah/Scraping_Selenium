```python
def filterXpath (xpath):
    for  counter_thr in range(len(xpath)):
        visible_txt = xpath[counter_thr]
        if len(visible_txt.xpath("@href")) > 0 :
            link = visible_txt.xpath("@href")[0]
            return link
```

```python
def validWebsite (web, link):
#     valid_link: 200 valid link that can be opened
#     error_link: 0 valid link could not be created
#     problem_web: 404 fundamental issue with the link 
    if type(link) == str:
        url_regex = re.compile ("(https?:?:\/{1,3})[a-z0-9%]|[a-z0-9.\-]+[.]")
        m=re.search (url_regex,link)
        check = ""
        try:
            if m is not None:
                check = link
            else:
                match = re.search ("\/{2,3}[a-z0-9%]|[a-z0-9.\-]+[.]",link)
                if match is not None:
                    check = "https::"+link
                else:
                    check = web+link
            driver = webdriver.Firefox(executable_path="C:\\Users\\Sarah\\geckodriver.exe")
#             driver.set_page_load_timeout(10)
            driver.get(website)
            driver.close()
            return (200,check)
        except:
            return (0,"")
    else:
        return (404,"")
```

```python
def aquiringPrivacyLink (page_content):
    link = ""
    tree = html.fromstring(page_content)
#             page_content[website]=driver.page_source
    match1 = re.search (">[\s]*([P|p]rivacy[ ]*[Policy|policy]*)[\s]*<",page_content) # Privacy
    match2 = re.search ('href="([\/\d\w]*[P|p]rivacy[\s_-]*[Policy|policy]*[\s]*)"',page_content) # Privacy
    print ("matches being found....")
    if match1 is not None:
        print ("regex match: ",match1.group(1))
        xpath_privacy = tree.xpath('//a[contains(text(),"{0}")]'.format(match1.group(1)))
        link = filterXpath (xpath_privacy)
    elif match2 is not None:
        print ("match 2: ",match2.group(1))
        link = match2.group(1)
    else:
        match3 = re.search (">[\s]*([T|t]erm[s]*[ ]*[of]*[ ]*[Use|use]*)[\s]*<",driver.page_source) # Term of Use
        if (match3 is not None):
            xpath_term = tree.xpath('//a[contains(text(),"{0}")]'.format(match3.group(1)))
            link = filterXpath (xpath_term)
            # re-scrape the TOS page
            page= driver.get(link)

            tree = html.fromstring(page.page_source)
            match1 = re.search ("([P|p]rivacy[ ]*[[P|p]olicy]*)",page.page_source) # Privacy
            if match1 is not None:
                xpath_privacy = tree.xpath('//a[contains(text(),"{0}")]'.format(match1.group(1)))
                link = filterXpath (xpath_privacy)
#             print (link,code, priv_linklink)
    return link
```

```python
def savingLink (error_link,except_link,privacy_link,page_content):
    [open ("error_link.txt","a",encoding="utf-8").write (w+"\n") for w in error_link]
    [open ("except_link.txt","a",encoding="utf-8").write (w+"\n") for w in except_link]
    with open('toplinks.csv', 'a',encoding="utf-8") as f:
        for w,l in privacy_link:
            writer = csv.writer(f)
            writer.writerow([w,l])
    with open('contentlinks.csv', 'a',encoding="utf-8") as f:
        for w,l in page_content.items():
            writer = csv.writer(f)
#             print (w,l)
            writer.writerow([w,l])
```
